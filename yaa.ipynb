```python
import matplotlib.pyplot as plt

# Data dari hasil analisis sebelumnya
riw_pe_labels = ['Tidak', 'PEB', 'PE', 'HELLP SYNDROM', 'Impending PE', 'Impending Eklamsia', 'Kejang Konvulsi', 'PE, HELLP Syndrome']
riw_pe_values = [526, 19, 6, 2, 2, 1, 1, 1]

riw_dm_labels = ['Tidak', 'Ya']
riw_dm_values = [556, 5]

# Membuat subplots untuk RIW PE dan RIW DM
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# Plot untuk RIW PE
axes[0].bar(riw_pe_labels, riw_pe_values, color='skyblue')
axes[0].set_title('Distribusi RIW PE')
axes[0].set_xlabel('RIW PE')
axes[0].set_ylabel('Jumlah Pasien')
axes[0].tick_params(axis='x', rotation=90)

# Plot untuk RIW DM
axes[1].bar(riw_dm_labels, riw_dm_values, color='lightgreen')
axes[1].set_title('Distribusi RIW DM')
axes[1].set_xlabel('RIW DM')
axes[1].set_ylabel('Jumlah Pasien')

# Menampilkan grafik
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data (sesuaikan dengan file Anda)
file_path = "C:\data mining\dataset\dataKasus-1.xlsx"
data = pd.read_excel(file_path, sheet_name='2022')

# Bersihkan data terkait Jarak Kelahiran dan Paritas
data_clean = data[['PARITAS', 'JARAK KELAHIRAN', 'RIW DM', 'RIW PE']].copy()

# Analisis Jarak Kelahiran dan Paritas untuk RIW DM
dm_grouped = data_clean.groupby(['RIW DM', 'JARAK KELAHIRAN'])['PARITAS'].value_counts().unstack().fillna(0)

# Analisis Jarak Kelahiran dan Paritas untuk RIW PE
pe_grouped = data_clean.groupby(['RIW PE', 'JARAK KELAHIRAN'])['PARITAS'].value_counts().unstack().fillna(0)

# Visualisasi untuk RIW DM
plt.figure(figsize=(10, 6))
sns.heatmap(dm_grouped, annot=True, cmap='Blues', fmt='.0f')
plt.title('Distribusi Paritas dan Jarak Kelahiran Berdasarkan RIW DM')
plt.ylabel('RIW DM dan Jarak Kelahiran')
plt.xlabel('Paritas')
plt.xticks(rotation=0)
plt.show()

# Visualisasi untuk RIW PE
plt.figure(figsize=(10, 6))
sns.heatmap(pe_grouped, annot=True, cmap='Greens', fmt='.0f')
plt.title('Distribusi Paritas dan Jarak Kelahiran Berdasarkan RIW PE')
plt.ylabel('RIW PE dan Jarak Kelahiran')
plt.xlabel('Paritas')
plt.xticks(rotation=0)
plt.show()
```
